```python
# Dynamic programming:
# Take row 11. See if cost is more if it is put in Group 1 or if it is placed in group 2 and take step accordingly
## Before that, writing the above in a function 

def group_cost_compute(in_mat,org_mat):
## Group cost 
    group_cost = 0
    for row in range(in_mat.shape[0]):
        for column in range(in_mat.shape[1]):
            if(org_mat.iloc[row,column] <> in_mat.iloc[row,column]):
                group_cost += 1
    return group_cost

def anon_group(in_mat):
    org_mat = in_mat.copy()
    df_vec = [0] * in_mat.shape[0]
    for i in range(0,in_mat.shape[0]):
        for j in range(0,in_mat.shape[0]):
            if j>=i:
                count = 0
            # Calculating jaccard similarity between two data rows i and j 
                for column in range(0,in_mat.shape[1]):    
                    if (in_mat.iloc[j,column] == 1 and in_mat.iloc[i,column] ==1):
                        count = count + 1
                df_vec[j] = df_vec[j] + count
    #print df_vec.index(max(df_vec))   ## Centroid
    row_max = df_vec.index(max(df_vec))
    center_row = in_mat.iloc[row_max,:]
    for rows in range(in_mat.shape[0]):
        in_mat.iloc[rows,:] = center_row  
    gr_cost = group_cost_compute(in_mat,org_mat)
    return in_mat,gr_cost  

#df_subset2 = df_all.iloc[11:20,:].copy()
#df_subset2b = df_all.iloc[0:11,:].copy()
```

```python
## DP example using Dictionaries, instead of lists

fibTable = {1:1, 2:1}
def fib(n):
   if n <= 2:
      return 1
   if n in fibTable:
      return fibTable[n]
   else:
      fibTable[n] = fib(n-1) + fib(n-2)
      return fibTable[n]
```
